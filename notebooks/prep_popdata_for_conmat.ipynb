```python
import pandas as pd

from tbh.demographic_tools import get_population_over_time
from tbh.runner_tools import DEFAULT_MODEL_CONFIG
from tbh.paths import DATA_FOLDER

# Load population data for the specified year
_, grouped_pop_df = get_population_over_time(DEFAULT_MODEL_CONFIG["iso3"], age_groups=DEFAULT_MODEL_CONFIG["age_groups"], scaling_factor=DEFAULT_MODEL_CONFIG["pop_scaling"])
year = 2025

pop_df = grouped_pop_df.loc[year]
```

```python
pop_df_pandas = pd.DataFrame({"lower.age.limit": pop_df.index, "population": pop_df.values})
```

```python
pop_df_pandas.to_csv(DATA_FOLDER / "Rscript" / f"KIR_pop_{year}.csv", index=False)
```
